```python
#타입 및 길이
print(type(doc1))
print(type(doc2))
print(len(doc1))
print(len(doc2))
```

```python
#konlpy를 이용한 명사 전처리(2글자 이상)
t = Okt()

tokens_doc1 = t.nouns(doc1)
tokens_doc2 = t.nouns(doc2)

tokens_doc1 = [token for token in tokens_doc1 if len(token) > 1]
tokens_doc2 = [token for token in tokens_doc2 if len(token) > 1]

print('doc1의 토큰 수 : ', len(tokens_doc1))
print('doc2의 토큰 수 : ', len(tokens_doc2))
print(tokens_doc1[:20])
print(tokens_doc2[:20])
```

```python
#단어별 빈도를 담을 딕셔너리를 계산하고 빈도 계산(doc1)
doc1_word_count = dict()

for word in tokens_doc1 : 
  doc1_word_count[word] = doc1_word_count.get(word,0) + 1

sorted_word_count_doc1 = sorted(doc1_word_count, key = doc1_word_count.get, reverse = True)

n = sorted_word_count_doc1[:20][::-1]
w = [doc1_word_count[key] for key in n]

import matplotlib.pyplot as plt

plt.figure(figsize = (10,10))
plt.barh(range(len(n)), w, tick_label = n)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.show()
```

```python
#단어별 빈도를 담을 딕셔너리를 계산하고 빈도 계산(doc2)
doc2_word_count = dict()

for word in tokens_doc2 : 
  doc2_word_count[word] = doc2_word_count.get(word,0) + 1

sorted_word_count_doc2 = sorted(doc2_word_count, key = doc2_word_count.get, reverse = True)

n = sorted_word_count_doc2[:20][::-1]
w = [doc2_word_count[key] for key in n]

import matplotlib.pyplot as plt

plt.figure(figsize = (10,10))
plt.barh(range(len(n)), w, tick_label = n)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.show()
```

###**토픽 모델링(Countvectorizer 이용)**

```python
#LDA를 위한 데이터 로드
import pandas as pd
doc1_lda = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data/1기결합.csv', encoding = 'cp949')

print(doc1_lda.shape)
doc1_lda.head()
```

```python
#LDA를 위한 데이터 로드
import pandas as pd
doc2_lda = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data/2기결합.csv', encoding = 'cp949')

print(doc2_lda.shape)
doc2_lda.head()
```

```python
#한글 형태소 분석기를 이용한 특성 벡터 변환
#konlpy의 형태소 분석기를 이용하여 토크나이저 함수 정의
#형태소 분석을 한 결과에서 명사, 동사, 형용사만을 반환하도록 함수 구현
from konlpy.tag import Okt

okt = Okt()

def my_tokenizer(doc) :
  return[token for token, pos in okt.pos(doc) if pos in ['Noun'] and len(token) > 1]

print('Result : ', my_tokenizer(doc1_lda.news[1]))
print('Result : ', my_tokenizer(doc2_lda.news[1]))
```

```python
#특성벡터 변환(doc1)
from sklearn.feature_extraction.text import CountVectorizer

doc_cv = CountVectorizer(max_features = 3000, tokenizer = my_tokenizer, min_df = 10, max_df = 0.5)

doc1_result = doc_cv.fit_transform(doc1_lda.news)

print(doc_cv.get_feature_names_out()[:10])
```

```python
#최적의 토픽 수 선택 함수(doc1)

def test_perplexity(cv, start = 10, end = 30, max_iter = 30, topic_word_prior = 0.1, doc_topic_prior = 1.0) : 
  iter_num = []
  per_value = []
  for i in range(start, end + 1):
    lda = LatentDirichletAllocation(n_components = i, max_iter = max_iter,
                                    topic_word_prior = topic_word_prior, doc_topic_prior= doc_topic_prior,
                                    learning_method = 'batch', n_jobs = -1, random_state = 2456)
    lda.fit(cv)
    iter_num.append(i)
    per_value.append(lda.perplexity(cv))
  plt.plot(iter_num,per_value, 'g-')
  plt.show()
  return start + per_value.index(min(per_value))

print("n_components with minimum perplexity : ", test_perplexity(doc1_result, start = 1, end = 15))
```

```python
#LDA 실행(doc1)
from sklearn.decomposition import LatentDirichletAllocation
import numpy as np
np.set_printoptions(precision = 3)

lda = LatentDirichletAllocation(n_components = 5, max_iter = 100, topic_word_prior=0.1, 
                                doc_topic_prior= 1.0, learning_method = 'batch',
                                n_jobs = -1, random_state = 1111)

review_topics_doc1 = lda.fit_transform(doc1_result)
```

```python
#토픽의 내용 출력

#토픽별로 비중이 높은 상위 n개의 단어를 출력하는 함수
def print_top_words(model, feature_names, n_top_words) : 
  for topic_idx, topic in enumerate(model.components_) : 
    print("Topic #%d : " % topic_idx, end = ' ')
    print(", ".join([feature_names[i] for i in topic.argsort()[:-n_top_words -1 : -1]])) #argsort 헷갈리니 잘 이해하기
  print()

print_top_words(lda, doc_cv.get_feature_names_out(), 10)
```
